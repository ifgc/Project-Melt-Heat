```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def remove_duplicate_rows(file_path, output_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Remove rows where all columns are identical
    df_deduplicated = df.drop_duplicates()
    
    # Save the cleaned DataFrame to a new CSV file
    df_deduplicated.to_csv(output_path, index=False)
    
    return df_deduplicated

# Specify the input and output file paths
input_file = 'LogFor_griRun/extracted_features.csv'
output_file = 'LogFor_griRun/cleaned_extracted_features.csv'

# Remove duplicate rows and save the cleaned DataFrame
cleaned_df = remove_duplicate_rows(input_file, output_file)

# Display the number of rows before and after cleaning
print(f"Number of rows before cleaning: {len(pd.read_csv(input_file))}")
print(f"Number of rows after cleaning: {len(cleaned_df)}")
```

```python
df1 = all_features
df2 = APO_ZTF_Photometery
df2 = df2.rename(columns={'oid': 'OID'})
display(df2.columns)
print(len(df1.columns),len(df2.columns))

print(df1['oid'].dtype,df2['OID'].dtype)


# Exclude 'OID' from columns_to_merge as it will be explicitly included in the merge
columns_to_merge = df2.columns[df2.columns != 'OID']

# Merge df2 columns to df1 based on the matching condition df1['oid'] == df2['OID']
MasterDF = pd.merge(df1, df2[['OID'] + list(columns_to_merge)], left_on='oid', right_on='OID', how='left')
display(MasterDF)
# Drop the redundant column 'OID' after merge
MasterDF.drop(columns=['OID'], inplace=True)
# Drop the redundant column 'OID' after merge
MasterDF.drop(columns=['ZTF_filtercode'], inplace=True)
display(MasterDF)
```
