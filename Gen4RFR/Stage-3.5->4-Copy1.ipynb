```python
df1 = all_features
df2 = APO_ZTF_Photometery
#this is patch for bug in stage 3 where oid and apogee id got swapped
df1 = df1.rename(columns={'apogee_id_01': 'oid','oid': 'apogee_id_01'})#bug in stage 3 has been reomved, on fuutre runs comment this out 

df2 = df2.rename(columns={'apogee_id_01': 'APOGEE_ID'})
df2 = df2.drop(columns={'oid','source_id_01'})
display(df2['APOGEE_ID'].dtype,df2['APOGEE_ID'].dtype)
display(df2.columns)
print(len(df1.columns),len(df2.columns))

# Exclude 'APOGEE_ID' from columns_to_merge as it will be explicitly included in the merge
columns_to_merge = df2.columns[df2.columns != 'APOGEE_ID']

# Merge df2 columns to df1 based on the matching condition df1['apogee_id_01'] == df2['APOGEE_ID']
MasterDF = pd.merge(df1, df2[['APOGEE_ID'] + list(columns_to_merge)], left_on='apogee_id_01', right_on='APOGEE_ID', how='left')

# Drop the redundant column 'APOGEE_ID' after merge
MasterDF.drop(columns=['APOGEE_ID'], inplace=True)
# Drop the redundant column 'APOGEE_ID' after merge
MasterDF.drop(columns=['ZTF_filtercode'], inplace=True)
display(MasterDF.head())
```

```python
counter = 0
for col in MasterDF.columns:
    print(f"[{counter}] {col}")
    counter = counter + 1
```

```python
# Plot for combined_light_curves
plt.figure(figsize=(8, 8))
plt.scatter(MasterDF['APOGEE_TEFF'], MasterDF['APOGEE_TEFF_SPEC'], s = 1, alpha=0.5, label='TEFF vs TEFF_SPEC')
plt.title('TEFF vs TEFF_SPEC - APOGEE')
plt.xlabel('TEFF')
plt.ylabel('TEFF_SPEC')
plt.legend()
plt.grid(True)
plt.show()

# Plot for combined_light_curves
plt.figure(figsize=(8, 8))
plt.scatter(MasterDF['APOGEE_TEFF_SPEC'], MasterDF['APOGEE_RV_TEFF'],  s = 1, color='red', alpha=0.5, label='RV_TEFF vs TEFF_SPEC')
plt.title('RV_TEFF vs TEFF_SPEC- APOGEE')
plt.xlabel('TEFF_SPEC')
plt.ylabel('RV_TEFF')
plt.legend()
plt.grid(True)
plt.show()

# Plot for combined_light_curves
plt.figure(figsize=(8, 8))
plt.scatter(MasterDF['APOGEE_TEFF'], MasterDF['APOGEE_RV_TEFF'], s = 1,color='green', alpha=0.5, label='TEFF vs RV_TEFF')
plt.title('TEFF vs RV_TEFF - APOGEE')
plt.xlabel('TEFF')
plt.ylabel('RV_TEFF')
plt.legend()
plt.grid(True)
plt.show()
```
