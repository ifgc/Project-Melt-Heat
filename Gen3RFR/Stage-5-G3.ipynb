```python
from astroquery.simbad import Simbad
import pandas as pd

# Customize Simbad to include specific columns for ugrizy magnitudes
custom_simbad = Simbad()
custom_simbad.TIMEOUT = 300
custom_simbad.add_votable_fields('flux(u)', 'flux(g)', 'flux(r)', 'flux(i)', 'flux(z)', 'flux(y)')

def get_ugrizy_magnitudes(gaia_edr3_id):
    """
    Retrieve ugrizy magnitudes from SIMBAD for a given Gaia EDR3 ID.

    Parameters:
    gaia_edr3_id (str): The Gaia EDR3 ID of the object.

    Returns:
    dict: A dictionary containing ugrizy magnitudes.
    """
    try:
        # Query SIMBAD with the Gaia EDR3 ID
        result = custom_simbad.query_object(f"Gaia EDR3 {gaia_edr3_id}")
        
        if result is None or len(result) == 0:
            print(f"No results found for Gaia EDR3 ID: {gaia_edr3_id}")
            return None
        
        # Extract ugrizy magnitudes
        magnitudes = {
            'u': result['FLUX_U'][0] if 'FLUX_U' in result.colnames else None,
            'g': result['FLUX_G'][0] if 'FLUX_G' in result.colnames else None,
            'r': result['FLUX_R'][0] if 'FLUX_R' in result.colnames else None,
            'i': result['FLUX_I'][0] if 'FLUX_I' in result.colnames else None,
            'z': result['FLUX_Z'][0] if 'FLUX_Z' in result.colnames else None,
            'y': result['FLUX_Y'][0] if 'FLUX_Y' in result.colnames else None
        }
        
        return magnitudes
    
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
gaia_edr3_id = '2364561950954447872'  # Replace with the actual Gaia EDR3 ID
magnitudes = get_ugrizy_magnitudes(gaia_edr3_id)

if magnitudes:
    # Convert to DataFrame for better visualization
    df = pd.DataFrame([magnitudes])
    print(df)
else:
    print("No magnitudes found or an error occurred.")
```
