## Read in ZTF Results and merge with APOGEE Data

```python
import pandas as pd
```

```python
ZTF_resultsColumns = pd.read_csv('ZTF_LightCurves/table_irsa_catalog_search_resultsApogees.csv')
APOGEECold = pd.read_csv('APOGEE_ColdData.csv')
display(ZTF_resultsColumns)
display(APOGEECold)
```

```python
def printAllCs(df):
    df = df
    for c in df.columns:
        print(f"'{c}',")
printAllCs(APOGEECold)
```

```python
def rename_list_with_prefix(columns, prefix):
    return [f"{prefix}_{col}" if not col.lower().startswith(prefix.lower()) and not col.lower().startswith("gaiaedr3") else col for col in columns]

# Define the specific columns you want to merge from df2
columns_to_merge = [
    'APOGEE_RA',
    'APOGEE_DEC',
    'TEFF_SPEC',
    'LOGG_SPEC',
    'TEFF',
    'TEFF_ERR',
    'LOGG',
    'LOGG_ERR',
    'RV_TEFF',
    'RV_LOGG',
    'RV_FEH',
    'GAIAEDR3_PARALLAX',
    'GAIAEDR3_PARALLAX_ERROR',
    'GAIAEDR3_PMRA',
    'GAIAEDR3_PMRA_ERROR',
    'GAIAEDR3_PMDEC',
    'GAIAEDR3_PMDEC_ERROR',
    'GAIAEDR3_PHOT_G_MEAN_MAG',
    'GAIAEDR3_PHOT_BP_MEAN_MAG',
    'GAIAEDR3_PHOT_RP_MEAN_MAG',
    'GAIAEDR3_R_MED_GEO',
    'GAIAEDR3_R_LO_GEO',
    'GAIAEDR3_R_HI_GEO',
    'GAIAEDR3_R_MED_PHOTOGEO',
    'GAIAEDR3_R_LO_PHOTOGEO',
    'GAIAEDR3_R_HI_PHOTOGEO',
    'FE_H',
    'FE_H_ERR',
    'FE_H_FLAG'
]

# Apply the renaming function to the list
columns_to_merge = rename_list_with_prefix(columns_to_merge, 'APOGEE')
```

```python
def rename_columns_with_prefix(df, prefix):
    return df.rename(columns={col: f"{prefix}_{col}" for col in df.columns 
                              if not (col.lower().startswith('ztf') or col.lower().startswith('apogee') or col.lower().startswith('gaiaedr3'))
                              and col not in ['oid', 'source_id_01', 'apogee_id_01']})



# Rename columns in ZTF_resultsColumns
ZTF_resultsColumns = rename_columns_with_prefix(ZTF_resultsColumns, 'ZTF')

# Rename columns in APOGEECold
APOGEECold = rename_columns_with_prefix(APOGEECold, 'APOGEE')
```

```python
df1= ZTF_resultsColumns
df2= APOGEECold
# Merge df2 columns to df1 based on the matching condition df1['c'] == df2['d']
MasterDF = pd.merge(df1, df2[['APOGEE_ID'] + columns_to_merge], left_on='apogee_id_01', right_on='APOGEE_ID', how='left')

# Drop the redundant column 'd' after merge
MasterDF.drop(columns=['APOGEE_ID'], inplace=True)
```

```python
display(MasterDF)
MasterDF.to_csv('MergedAPOGEEZTF_Data.csv', index=False)
```
